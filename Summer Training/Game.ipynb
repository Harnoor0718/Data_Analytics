```python
team_runs = df.groupby('match_won_by')['runs_total'].sum()
plt.figure(figsize=(8, 5))
team_runs.plot(kind='bar', color='skyblue')
plt.title('Total Runs Scored by Teams in Matches Won')
plt.xlabel('Teams')
plt.ylabel('Total Runs')
plt.show()
```

```python
team_wins = df['match_won_by'].value_counts()
top_5_teams = team_wins.head(5)
plt.figure(figsize=(6, 6))
top_5_teams.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Top 5 Teams with Most Matches Won')
plt.ylabel('')
plt.show()
```
