```python
plt.figure(figsize=(8, 5))
df[['Duration', 'Pulse', 'Maxpulse', 'Calories']].boxplot()
plt.title('Box Plot: Outlier Detection')
plt.ylabel('Values')
plt.show()
```

```python
def detect_outliers(column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return len(outliers)
outlier_counts = {col: detect_outliers(col) for col in ['Duration', 'Pulse', 'Maxpulse', 'Calories']}
print("Outlier counts:", outlier_counts)
```

```python
Q1 = df['Calories'].quantile(0.25)
Q3 = df['Calories'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_cleaned = df[(df['Calories'] >= lower_bound) & (df['Calories'] <= upper_bound)]
plt.figure(figsize=(10, 5))

# Original Histogram
plt.subplot(1, 2, 1)
plt.hist(df['Calories'], bins=20, color='skyblue', alpha=0.7)
plt.title('Original Calories Histogram')
plt.xlabel('Calories')
plt.ylabel('Frequency')

# Cleaned Histogram
plt.subplot(1, 2, 2)
plt.hist(df_cleaned['Calories'], bins=20, color='green', alpha=0.7)
plt.title('Cleaned Calories Histogram')
plt.xlabel('Calories')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
