```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('loan.csv')
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset["ApplicantIncome"].fillna(dataset["ApplicantIncome"].mean(), inplace=True)
```

```python
sns.distplot(dataset["ApplicantIncome"])
plt.show()
```

```python
dataset.describe()
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(dataset[["ApplicantIncome"]])
```

```python
dataset["ApplicantIncome__ss"] = pd.DataFrame(ss.transform(dataset[["ApplicantIncome"]]), columns = ["x"])
```

```python
dataset.head(3)
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(12, 5)) 
# Plot before transformation
plt.subplot(1, 2, 1) 
plt.title("Before") 
sns.distplot(dataset["ApplicantIncome"]) 

# Plot after transformation 
plt.subplot(1, 2, 2) 
plt.title("After") 
sns.distplot(dataset["ApplicantIncome_ss"]) 
plt.show()
```
