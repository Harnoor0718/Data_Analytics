```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("placement-dataset.csv")
dataset.head(3)
```

```python
dataset.drop(dataset.columns[0], axis=1, inplace=True)
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x="cgpa", y = "iq", data = dataset, hue="placement")
```

```python
x = dataset[['cgpa','iq']]
y = dataset["placement"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
print(x_train.isnull().sum())
print(y_train.isnull().sum())
```

```python
print(x_train.shape)  # (80, 2)
print(y_train.shape)  # (80,)
print(x_test.shape)   # (20, 2)
print(y_test.shape)   # (20,)
```

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100
```

```python
lr.predict([[8.14, 6.52]])
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf = lr)
```
