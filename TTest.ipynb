```python
Before = np.array([50, 60, 45, 55, 55, 70, 40, 75, 80, 65, 70, 60, 55, 45, 75, 60, 50, 65, 70 ])
After =  np.array([60, 70, 55, 75, 65, 80, 90, 70, 75, 65, 55, 60, 80, 60, 65, 55, 70, 75])
```

```python
std_a = np.std(After)
std_b = np.std(Before)
mean_a = np.mean(After)
mean_b = np.mean(Before)
```

```python
t_cal = (mean_a-mean_b )/np.sqrt(((std_a*std_a)/len(After)))+((std_b*std_b))
t_cal
```
