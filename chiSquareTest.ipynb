```python
row1 = np.array([40,45,25,10])
row2 = np.array([35,30,20,30])
```

```python
sum_r1 = np.sum(row1)
sum_r2 = np.sum(row2)
sum_row = np.array([sum_r1, sum_r2])
sum_row
```

```python
sum_col = row1 + row2
sum_col
```

```python
exp = []
for i in sum_row:
    for j in sum_col:
        value = (i*j)/235
        exp.append(value)
    
```

```python
obj  = np.array([40,45,25,10,35,30,20,30])
```

```python
np.sum((np.square(obj - exp)) / exp)
```
