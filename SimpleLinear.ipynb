```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
dataset = pd.read_csv('placement.csv')
dataset.head(3)
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x='cgpa', y='package', data=dataset)
plt.show()
```

```python
dataset.isnull().sum()
```

```python
x = dataset[['cgpa']]
x.ndim
y = dataset['package']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
# y = m*x + c
```

```python
lr.coef_   # m value
```

```python
lr.intercept_   # c value
```

```python
# y = 0.56952601*x + 0.9945844024642021
```

```python
# Testing the model
y_pred = lr.predict([[6.89]])
y_pred
```

```python
# Checking the accuracy score
accuracy = lr.score(x_test, y_test)*100
accuracy # Poor accuracy due to less data
```

```python
y_pred = lr.predict(x)
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x='cgpa', y='package', data=dataset)
plt.plot(dataset["cgpa"],y_pred, color='red' )
plt.legend(['org data' ,'predicted line'])
plt.savefig('predict.jpg')
plt.show()
```
